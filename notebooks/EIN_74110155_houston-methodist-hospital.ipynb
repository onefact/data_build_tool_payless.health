`````python
import subprocess
from pathlib import Path

def convert_ashx_to_json_by_key(ashx_file, key, output_dir):
    tmp_file_path = output_dir / f'{key}.tmp'
    print('input:', ashx_file)
    print('tmp_file', tmp_file_path)
    with open(tmp_file_path, 'w') as tmp_file:
        subprocess.run(['jq', f'.["{key}"][]', str(ashx_file)], stdout=tmp_file)
    json_file = output_dir / f'{key}.json'
    with open(json_file, 'w') as json_file:
        subprocess.run(['jq', '-s', '.', str(tmp_file_path)], stdout=json_file)
    print('output:', json_file)

`````

`````python
%%sql 
COPY (
WITH max_min_charges AS (
  SELECT
    billing_code, 
    description,
    MIN(negotiated_rate) AS min_charge,
    MAX(negotiated_rate) AS max_charge
  FROM '/Users/me/data/payless_health/EIN_74110155_houston-methodist-hospital/*.parquet'
  GROUP BY billing_code, description
),

min_max_names AS (
  SELECT
    mmc.billing_code,
    mmc.description,
    mmc.min_charge, 
    mmc.max_charge,
    FIRST_VALUE(payer) OVER (PARTITION BY mmc.billing_code, mmc.description ORDER BY negotiated_rate ASC) AS name_minimum,  
    FIRST_VALUE(payer) OVER (PARTITION BY mmc.billing_code, mmc.description ORDER BY negotiated_rate DESC) AS name_maximum
  FROM max_min_charges mmc
  JOIN '/Users/me/data/payless_health/EIN_74110155_houston-methodist-hospital/*.parquet' sc
  ON mmc.billing_code = sc.billing_code AND mmc.description = sc.description
)

SELECT DISTINCT
  billing_code,
  description,
  min_charge,
  max_charge,
  name_minimum,
  name_maximum
FROM min_max_names
) TO '/Users/me/data/payless_health/EIN_74110155_houston-methodist-hospital_analysis.parquet' (COMPRESSION ZSTD);
`````
