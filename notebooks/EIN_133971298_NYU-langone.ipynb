`````python
import csv
from collections import defaultdict

counts = defaultdict(int)

header_row = 2

with open('/tmp/download', 'r') as infile:
    reader = csv.reader(infile)
    for _ in range(header_row):
      next(reader)
    header = next(reader)
    
    new_header = []
    for col in header:
        if col in counts:
            counts[col] += 1
            new_header.append(f"{col} ({counts[col]})")
        else:
            counts[col] = 0
            new_header.append(col)
            
    with open('/tmp/download-no-duplicates', 'w') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(new_header)
        
        for row in reader:
            writer.writerow(row)
`````

> please write a python script to read a csv and write a new csv after removing any non-numeric values using the regex '[^0-9.]' starting on the third row. please only apply the regex to values in the csv for every column except the first two



`````python
import csv
import re

input_file = '/tmp/download-no-duplicates'
output_file = '/tmp/download-no-duplicates-replaced' 

with open(input_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    rows = list(reader)

header = rows[0]

for row in rows[1:]:
    for i in range(3, len(row)):
        # replace all non-numeric characters with an empty string
        row[i] = re.sub('[^0-9\.\-]', '', row[i])
        if row[i] == '' or row[i] == '-1.00':
            row[i] = 'NC'
        

with open(output_file, 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    writer.writerows(rows[1:])
`````
