# Goal - dropdown bar chart for prices by codes for the 18 codes that are matched in the top 25

St Lukes will have more prices listed. Lehigh has 2; the minimum and the maximum.

Sorted bars by price, descending.

Selection of code: Code - description
Table: hospital name - payr - price

Export parquet file in this format

`````python
df_stlukes_matches_filtered = df_stlukes_matches[['cpt_drg', 'payer', 'charge', 'description']]
df_stlukes_matches_filtered['hospital'] = 'St. Luke\'s'
df_stlukes_matches_filtered['payer'] = 'St. Luke\'s - ' + df_stlukes_matches['payer'].str.capitalize()
`````

`````python
df_stlukes_matches_filtered
`````

`````python
code_to_description = dict(zip(df_stlukes_matches_filtered['cpt_drg'].values, df_stlukes_matches_filtered['description'].str.lower().str.capitalize().values))
`````

`````python
df_lehigh_matches_melted = df_lehigh_matches[['billing_code', 'Maximum_IP', 'Minimum_IP']].melt(id_vars=['billing_code'], var_name='payer', value_name='charge').rename(columns={'billing_code': 'cpt_drg'})
df_lehigh_matches_melted['payer'] = 'Lehigh Valley - ' + df_lehigh_matches_melted['payer']
df_lehigh_matches_melted['hospital'] = 'Lehigh Valley'
`````

`````python
df_lehigh_matches_melted
`````

`````python
df_final = pd.concat([df_lehigh_matches_melted, df_stlukes_matches_filtered])
`````

`````python
df_final.rename(columns={'cpt_drg': 'Billing_Code', 'payer': 'Hospital_and_Payor', 'charge': 'Charge', 'hospital': 'Hospital', 'description': 'Description'}, inplace=True)
`````

`````python
df_final_matches = df_final[df_final['Billing_Code'].isin(top_codes_matches)]
`````

`````python
df_final_matches['Billing_Code_and_Description'] = df_final_matches['Billing_Code'] + ' - ' + df_final_matches['Billing_Code'].map(code_to_description)
`````

`````python
df_final_matches
`````
