https://www.cms.gov/medicare/payment/prospective-payment-systems/acute-inpatient-pps/ms-drg-classifications-and-software 

ICD-10 MS-DRG Definitions Manual Files V41 (ZIP): A zip file with the ICD-10 MS DRG Definitions Manual (Text Version) contains the complete documentation of the ICD-10 MS-DRG Grouper logic. - Updated 8/22/2023:

https://www.cms.gov/files/zip/icd-10-ms-drg-definitions-manual-files-v41.zip

The structure of the files in the zip file is as follows:

* mdcs_00_07.txt: MDCs 00-07
* mdcs_08_11.txt: MDCs 08-11
* mdcs_12_21.txt: MDCs 12-21
* mdcs_22_25.txt: MDCs 22-25

MDC refers to Major Diagnostic Category.  There are 25 MDCs that each relate to the International Classification of Diseases developed by the World Health Organization.  

Each MDC is divided into DRGs.  There are many DRGs.  Each DRG is divided into MS-DRGs.  There are many MS-DRGs.  Each MS-DRG is divided into MS-DRG Severity Levels.  There are 4 MS-DRG Severity Levels.

To relate the major diagnostic category to every DRG code, we need to parse the text files in the zip file.

We isolate the text following "MDC" and then isolate the text following "DRG".

# Parse the DRG and MDC mapping

Prompt:

> please write a python function to read the first three characters of every line as the first column, separated by a space, the next two characters to be the value of the second column, separated by a space, a one-character third column value, a space, and the rest of the line being the value of the fourth column



```python
def parse_lines(lines):
  result = []
  for line in lines:
    # First 3 characters are the DRG code
    col1 = line[0:3]
    # One space separates the DRG code from the two-character MDC code
    col2 = line[4:6]
    # Two spaces separate the MDC code from the one-character MS code
    col3 = line[8:9]
    # Two spaces separate the MS code from the description
    col4 = line[11:]
    result.append((col1, col2, col3, col4.rstrip()))
  return result

with open('/Users/me/data/payless_health/icd-10-ms-drg-definitions-manual-files-v41/appendix_A.txt') as f:
    lines = f.readlines()
    parsed = parse_lines(lines[11:])
```

```python
import pandas as pd
df = pd.DataFrame(parsed, columns=['DRG', 'MDC', 'MS', 'Description'])
```
